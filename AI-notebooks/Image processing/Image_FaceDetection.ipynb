##**FaceDetection**
Implement a solution to detect faces from an input image using a pre-built model
such as Haar cascade detector.

#Implementation
In this task, we will use the Haar cascade face detector to find faces from an input
image.With the
advance of modern machine learning technologies, many other models for face
detection have been developed. But the Haar cascade detector is still one of the
fastest detectors though it may not be the most accurate one.


First we install the OpenCV package and import it. We also need to import the
patches for imshow() in OpenCV as it does not work properly within the Google.

Although an alternative could be Matplotlib also.

we have downloaded "haarcascade" to identify the face. This is the crux of our solution. 

```
!pip install opencv-python
import cv2
from google.colab.patches import cv2_imshow
```

```
# Load the face detection model from the model file
!wget 'https://github.com/yongminli/data/raw/main/haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```
# Load the model file and image files to Google Colab
from google.colab import files
file = files.upload()
```

```
x = "Papel-de-parede-Witcher-3-4k-1.webp"
img2 = cv2.imread(x)
cv2_imshow(img2)
```

```
img = cv2.imread('Papel-de-parede-Witcher-3-4k-3.webp')
cv2_imshow(img)
```

```

gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Detect faces
# faces = face_cascade.detectMultiScale(image=gray, scaleFactor=1.2, minNeighbors=5, minSize=(50,50))
faces = face_cascade.detectMultiScale(image=gray, minNeighbors=5)

# Display the detecition results
for (x, y, w, h) in faces:
    cv2.rectangle(img2, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img2)
```

```
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Detect faces
# faces = face_cascade.detectMultiScale(image=gray, scaleFactor=1.2, minNeighbors=5, minSize=(50,50))
faces = face_cascade.detectMultiScale(image=gray2, minNeighbors=5)

# Display the detecition results
for (x, y, w, h) in faces:
    cv2.rectangle(img2, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img2)
```

```
t_lower = 60 
t_upper = 150
  
# Applying the Canny Edge filter
edge = cv2.Canny(gray, t_lower, t_upper)
cv2_imshow(edge)
```
