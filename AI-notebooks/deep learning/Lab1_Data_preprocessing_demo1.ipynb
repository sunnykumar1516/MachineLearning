# Load data using pandas package
Data structure and data visualiazation


```python
df = pd.read_csv('/content/drive/MyDrive/lab_code/IowaHousingPrices.csv')  #Load data with the corresponding path in google drive 
```

```python
print (df.describe()) # Display some statistical results
```

```python
X = df['SquareFeet'].values.reshape(-1,1) # Transform the dataframe dataset into an array with a shape of (1460,1), in order to satify the requirements of LinearRegression model.
Y = df['SalePrice'].values.reshape(-1,1)   

# Visualization
plt.scatter(X, Y, s=20, marker='o', edgecolor=['green'], c='none')  # Scatter data 
plt.xlabel('SquareFeet', fontsize=20)
plt.ylabel('SalePrice', fontsize=20)
plt.show()
```

# Model training

```python
reg = LinearRegression().fit(X, Y)              # Model fitting
print (reg.coef_)                               # Weight of linear regression
print (reg.intercept_)                          # Bias of linear regression
```

```python
fitted_train = reg.predict(X)    # Predicted the training set

# Visualization
plt.scatter(X, Y, s=20, marker='o', edgecolor=['green'], c='none', label='Actual prices')
plt.scatter(X, fitted_train, s=20, marker='o', edgecolor=['red'], c='none', label='Fitted prices')
plt.xlabel('SquareFeet', fontsize=20)
plt.ylabel('SalePrice', fontsize=20)
plt.legend(fontsize=15)
plt.show()
```

# Model testing and a price prediction

```python
X_test = np.asarray([3000]).reshape(-1,1) # A test SquareFeet of 3000, and transform its shape to be (1,1)
print (reg.predict(X_test))
```
